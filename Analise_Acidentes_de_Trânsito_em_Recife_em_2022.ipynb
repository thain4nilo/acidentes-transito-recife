# Transformação de Dados

### Removendo colunas que não estão preenchidas ou não tem utilidade

```python
#Remover colunas que não estão preenchidas ou sem utilidade para o foco da análise [HORA, PROTOCOLO, NUMERO, DETALHE, DIVISAO123]
acidentes_df = acidentes.drop(["Protocolo","hora","divisao_via1", "divisao_via2","divisao_via3","numero","detalhe_endereco_acidente","num_semaforo"],axis=1)
#acidentes_df.head(5)
```

### Ajustando formato das informações das colunas

```python
# convertendo data para formato Date
acidentes_df['data']= pd.to_datetime(acidentes_df['data'], format='%Y-%m-%d')

# Ajustando tipo das colunas que são inteiro e estavam como string
acidentes_df['vitimasfatais'] = acidentes_df['vitimasfatais'].str.replace(",0", "").astype(int)
acidentes_df['vitimas'] = acidentes_df['vitimas'].str.replace(",0", "").astype(int)
acidentes_df['auto'] = acidentes_df['auto'].str.replace(",0", "").astype(int)
acidentes_df['moto'] = acidentes_df['moto'].str.replace(",0", "").astype(int)
acidentes_df['ciclom'] = acidentes_df['ciclom'].str.replace(",0", "").astype(int)
acidentes_df['ciclista'] = acidentes_df['ciclista'].str.replace(",0", "").astype(int)
acidentes_df['pedestre'] = acidentes_df['pedestre'].str.replace(",0", "").astype(int)
acidentes_df['onibus'] = acidentes_df['onibus'].str.replace(",0", "").astype(int)
acidentes_df['caminhao'] = acidentes_df['caminhao'].str.replace(",0", "").astype(int)
acidentes_df['viatura'] = acidentes_df['viatura'].str.replace(",0", "").astype(int)
acidentes_df['velocidade_max_via']=acidentes_df['velocidade_max_via'].str.replace("[A-Za-z/ ]","")

# Agrupando tipos de acidentes, devido a pouca ocorrencia dos mesmos e para simplicação das categorias
acidentes_df.loc[((acidentes_df['tipo']=="RENDIÇÃO")|
                   (acidentes_df['tipo']=="PROTESTO")|
                   (acidentes_df['tipo']=="TOMBAMENTO")|
                   (acidentes_df['tipo']=="QUEDA")|
                   (acidentes_df['tipo']=="CAPOTAMENTO")|
                   (acidentes_df['tipo']=="ATROPELAMENTO DE ANIMAL")|
                   (acidentes_df['tipo']=="ENGAVETAMENTO")|
                   (acidentes_df['tipo']=="APOIO EMLURB")|
                   (acidentes_df['tipo']=="MONITORAMENTO")|
                   (acidentes_df['tipo']=="FISCALIZAÇÃO")),'tipo'] = 'OUTROS'

acidentes_df.loc[((acidentes_df['tipo']=="COLISÃO FRONTAL")|
                   (acidentes_df['tipo']=="COLISÃO TRANSVERSAL")|
                   (acidentes_df['tipo']=="COLISÃO TRASEIRA")|
                   (acidentes_df['tipo']=="CHOQUE")|
                   (acidentes_df['tipo']=="COLISÃO LATERAL")), 'tipo'] = 'COLISÃO'

acidentes_df
```

```python
acidentes_df['velocidade_max_via'].value_counts()
```

### Adicionando colunas que são interessantes para a análise

```python
# COLUNAS REFERENTES A DATA
# usar o atributo .dt para acessar componentes específicos da data, onde %B é um código de formatação que representa o nome completo do mês e %m representa o número do mes
acidentes_df['mes'] = acidentes_df['data'].dt.strftime('%m') + "_" + acidentes_df['data'].dt.strftime('%B')
# usar o atributo .dt para acessar componentes específicos da data, onde %a é um código de formatação que representa o dia da semana e %w o numero do dia da semana
acidentes_df['dia_da_semana'] = acidentes_df['data'].dt.strftime('%w') + "_" + acidentes_df['data'].dt.strftime('%A')
```

## Existe relação entre o período do ano/data comemorativa e aumento de acidentes?

```python
# contando quantos acidentes tem que cada mes e colocando em um dataframe
acidentes_mes = acidentes_df['mes'].value_counts().reset_index(name='Qntd_Acidentes')

# formatando o dataframe
acidentes_mes = acidentes_mes.rename(columns={'index': 'mes'}) #renomeando coluna mês que perdeu o nome
acidentes_mes = acidentes_mes.sort_values("mes") #ordenando pelo mês

# exibir tabela
display(acidentes_mes)

# plotar grafico de linhas
display(px.line(acidentes_mes, x="mes", y="Qntd_Acidentes", markers=True))
```

**Análise**

O segundo semestre do ano concentra o maior número de acidentes em Recife no ano passado. Algumas hipóteses que podem estar relacionadas são:

*   Metade do ano com maior número de feriados
*   Semestre de volta à normalidade pós-pandemia

Obs: Para analisar se este é um comportamento padrão ao longo dos anos, seria necessário avaliar anos anteriores.

## Há alguma relação do dia da semana com a natureza do acidente (sem vitima / com vitima / com vitima fatal)?

```python
acidentes_natureza = acidentes_df[['dia_da_semana','natureza']].groupby('dia_da_semana').value_counts().reset_index(name="Qnt")

display(px.pie(acidentes_natureza.sort_values('dia_da_semana'), names="dia_da_semana", values="Qnt",facet_col="natureza"))

display(px.histogram(acidentes_natureza.sort_values('dia_da_semana'),x="dia_da_semana", y="Qnt", color="natureza",
             barnorm = "percent",
             text_auto= True))
```

**Análise**

Não verificado nenhuma relação para as naturezas sem vitima e com vítima, contudo podemos verificar que segunda, quarta e domingo apresentam maior incidencia de acidentes fatais.
* 1,4% dos acidentes na segunda-feira foram fatais
* 1,3% dos acidentes na quarta-feira foram fatais
* 1,6% dos acidentes no domingo foram fatais.

Esta informação com relação ao domingo se mostra relevante, dado que é o dia com menos acidentes registrados, mas apresenta o maior indice de fatalidade proporcional.



## Como podemos distribuir os guardas de transito nos bairros durante a semana?

```python
# verificamos a distribuição de acidentes pelos bairros durante a semana
acidentes_bairro = acidentes_df[['dia_da_semana','bairro']].value_counts().reset_index(name='Qnt')

# há muitos bairros com um numero de acidentes não significativo
px.histogram(acidentes_bairro,x='Qnt',color='bairro')
```

Considerando que um ano tem em torno de **52 semanas**, para que haja relevancia neste dado, podemos considerar que haja acidente em pelo menos **1/3 do ano** naquele bairro, o que significa dizer que em algum dia da semana o bairro atingiu um total de no **mínimo de 17 acidentes**.

```python
filtro_bairros = acidentes_bairro[acidentes_bairro['Qnt']>=17] #filtrando bairros que tenham tido mais de 17 ocorrencias no ano em algum dia da semana
bairros_interesse = filtro_bairros['bairro'].unique() #criando lista destes bairros, sem duplicata

acidentes_bairro = acidentes_bairro[acidentes_bairro['bairro'].isin(bairros_interesse)] #filtrando no df de acidentes_bairro

px.scatter(acidentes_bairro.sort_values('dia_da_semana'),x='dia_da_semana', y='Qnt',color='bairro')
```

**Analise**
Com este resultados seria possível alocar a CTTU de acordo com a quantidade de recurso pesssoal disponível nos bairros que mais se destacam (se afastam) dos demais considerando o eixo y.

Como um exemplo, se houver apenas 4 equipes disponíveis, uma forma de alocação seria:
* **segunda:** 2 em Boa viagem / 1 em Santo Amaro / 1 na Imbiribeira
* **terça:** 2 em Boa viagem / 1 na Imbiribeira / 1 no Cordeiro

## Como podemos alocar as ambulâncias de atendimento de emergencia (SAMU, BOMBEIROS) nos bairros durante a semana?

```python
# filtando apenas os acidentes que apresentaram vitima
acidentes_bairro_natureza = acidentes_df[acidentes_df['natureza']!="SEM VÍTIMA"]

# agrupando por bairro
acidentes_socorro = acidentes_bairro_natureza[['dia_da_semana','bairro']].value_counts().reset_index(name='Qnt')

# filtrando o universo de interesse
acidentes_socorro = acidentes_socorro[acidentes_socorro['bairro'].isin(bairros_interesse)] #filtrando no df de acidentes_bairro

px.scatter(acidentes_socorro.sort_values("dia_da_semana"),x='dia_da_semana', y='Qnt',color='bairro')
```

**Analise**
A analise é a mesma do item anterior, contudo há uma mudança dos bairros de acordo o envolvimento de vitimas nos acidentes. Esta visão pode se aproximar ou se afastar da visão anterior a partir da proporção da natureza dos acidentes de cada bairro.

Utilizando o mesmo exemplo de apenas 4 equipes do SAMU disponíveis, uma forma de alocação seria:
* **segunda:** 2 em Boa viagem / 1 em Santo Amaro / 1 no IPSEP
* **terça:** 2 em Boa viagem / 1 na Imbiribeira / 1 no Cordeiro
